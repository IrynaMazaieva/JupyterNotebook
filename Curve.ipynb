```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
sym.init_printing()
t = sym.Symbol('t')
print("r(t) = {x(t), y(t), z(t)}")

x = sym.simplify(input('x(t) '))
y = sym.simplify(input('y(t) '))
z = sym.simplify(input('z(t) '))
(int_t1, int_t2) = (float(input('t: from ')), float(input('to '))) 
x_array, y_array, z_array = [], [], []
r = [x, y, z]
t1 = min(int_t1, int_t2)
t2 = max(int_t1, int_t2)
i = t1
while i < t2:
    x_array.append(x.subs('t', i))
    y_array.append(y.subs('t', i))
    z_array.append(z.subs('t', i))
    i += 0.01
    
print(r)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z_array, label='parametric curve')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

t = float(input('t = '))

r_t = [sym.diff(r[0], 't'), sym.diff(r[1], 't'), sym.diff(r[2], 't')]
r_tt = [sym.diff(r[0], 't', 2), sym.diff(r[1], 't', 2), sym.diff(r[2], 't', 2)]
r_ttt = [sym.diff(r[0], 't', 3), sym.diff(r[1], 't', 3), sym.diff(r[2], 't', 3)]

#print(r_t, r_tt, r_ttt)

r_in_t = [r[0].subs('t', t), r[1].subs('t', t), r[2].subs('t', t)]
r_t_in_t = [r_t[0].subs('t', t), r_t[1].subs('t', t), r_t[2].subs('t', t)]
r_tt_in_t = [r_tt[0].subs('t', t), r_tt[1].subs('t', t), r_tt[2].subs('t', t)]
r_ttt_in_t = [r_ttt[0].subs('t', t), r_ttt[1].subs('t', t), r_ttt[2].subs('t', t)]

r_t_mod = sym.sqrt((r_t_in_t[0])**2 + (r_t_in_t[1])**2 + (r_t_in_t[2])**2).evalf()
r_tt_mod = sym.sqrt((r_tt_in_t[0])**2 + (r_tt_in_t[1])**2 + (r_tt_in_t[2])**2).evalf()
r_ttt_mod = sym.sqrt((r_ttt_in_t[0])**2 + (r_ttt_in_t[1])**2 + (r_ttt_in_t[2])**2).evalf()

r_t_dot_r_tt = [r_t[1]*r_tt[2] - r_t[2]*r_tt[1], -r_t[0]*r_tt[2] + r_tt[0]*r_t[2], r_t[0]*r_tt[1] - r_tt[0]*r_t[1]]
r_t_dot_r_tt_in_t = [r_t_dot_r_tt[0].subs('t', t), r_t_dot_r_tt[1].subs('t', t), r_t_dot_r_tt[2].subs('t', t)]
r_t_dot_r_tt_in_t_mod = sym.sqrt((r_t_dot_r_tt_in_t[0])**2 + (r_t_dot_r_tt_in_t[1])**2 + (r_t_dot_r_tt_in_t[2])**2).evalf()

tau = [r_t_in_t[0]/r_t_mod, r_t_in_t[1]/r_t_mod, r_t_in_t[2]/r_t_mod]
beta = [r_t_dot_r_tt_in_t[0]/r_t_dot_r_tt_in_t_mod, r_t_dot_r_tt_in_t[1]/r_t_dot_r_tt_in_t_mod, r_t_dot_r_tt_in_t[2]/r_t_dot_r_tt_in_t_mod]

beta_dot_tau = [beta[1]*tau[2] - beta[2]*tau[1], -beta[0]*tau[2] + tau[0]*beta[2], beta[0]*tau[1] - tau[0]*beta[1]]
beta_dot_tau_mod = sym.sqrt((beta_dot_tau[0])**2 + (beta_dot_tau[1])**2 + (beta_dot_tau[2])**2).evalf()

new = [beta_dot_tau[0]/beta_dot_tau_mod, beta_dot_tau[1]/beta_dot_tau_mod, beta_dot_tau[2]/beta_dot_tau_mod]

print('  - ', tau)
print('  - ', beta)
print('   - ', new)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z_array, label='parametric curve')
ax.quiver(r_in_t[0], r_in_t[1], r_in_t[2], tau[0], tau[1], tau[2], length=1.0, label='tau', color='black')
ax.quiver(r_in_t[0], r_in_t[1], r_in_t[2], beta[0], beta[1], beta[2], length=1.0, label='tau', color='red')
ax.quiver(r_in_t[0], r_in_t[1], r_in_t[2], new[0], new[1], new[2], length=1.0, label='tau', color='yellow')
plt.show()
```
